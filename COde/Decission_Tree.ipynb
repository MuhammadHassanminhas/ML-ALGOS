```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris_data = load_iris()
data = pd.DataFrame(data=iris_data.data , columns=iris_data.feature_names)
```

```python
data
```

```python
data['target'] = iris_data.target
```

```python
data
```

```python
dr = DecisionTreeClassifier()
dr_temp = DecisionTreeClassifier(max_depth=13)
data.columns
```

```python
from sklearn.model_selection import train_test_split
X = data[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']]
```

```python
Y = data['target']
```

```python
X_train , X_test , Y_train ,Y_test = train_test_split(X,Y , random_state=42 , test_size=0.2)
```

```python
dr.fit(X_train ,Y_train)
```

```python
dr_temp.fit(X_train , Y_train)
```

```python
Y_pred = dr.predict(X_test)
```

```python
Y_pred_temp = dr_temp.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , precision_score
print(confusion_matrix(Y_test , Y_pred))
print(accuracy_score(Y_test , Y_pred))
print(confusion_matrix(Y_test , Y_pred_temp))
```
